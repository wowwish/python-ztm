```python
df1 = pd.DataFrame(data_frame, columns = ['Name', 'Wage', 'Value']) # Create a new filtered data frame from the old data frame

# Convert Wage and Value to actual numbers and find difference
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

value = df1['Value'].replace('[\€,]', '', regex=True).apply(value_to_float)
wage = df1['Wage'].replace('[\€,]', '', regex=True).apply(value_to_float)

df1['Wage'] = wage
df1['Value'] = value

# Find players with the highest difference between Values and Wages (Underdogs)
df1['difference'] = df1['Value'] - df1['Wage']
df1.sort_values('difference', ascending=False) # sort by difference in Descending order
```

```python
import seaborn as sns
sns.set() # set default options for plots from seaborn

graph = sns.scatterplot(x='Wage', y='Value', data=df1) # Scatter Plot of Wage and Value from data frame df1
```

```python
# REFER bokeh DOCUMENTATION: https://docs.bokeh.org/en/latest/docs/user_guide/tools.html?highlight=hover
from bokeh.plotting import figure,show

TOOLTIPS = [
    ('index', '$index'),
    ('(Wage, Value)', '(@Wage, @Value)'),
    ('Name', '@Name')
]

p = figure(title='Soccer 2019', x_axis_label='Wage', y_axis_label='Value', width=700, height=700, tooltips=TOOLTIPS)
p.circle('Wage', 'Value', size=10, source=df1)
show(p)
```
